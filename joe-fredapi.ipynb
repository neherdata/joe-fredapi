```python
base_series_url = 'https://api.stlouisfed.org/fred/series/observations?'

series_id = 'PCOPPUSDM'
realtime_start = '1990-01-01'
observation_start = '1990-01-01'
series_url = f"{base_series_url}series_id={series_id}&api_key={api_key}&file_type={file_type}&observation_start={observation_start}"

series_request = requests.get(series_url)
series_request
```

```python
obsdict = series_request.json()
obsdict.keys()
```

```python
cols = []
cols = list(obsdict['observations'][0].keys())
cols
```

```python
realtime_start = []
realtime_end = []
date = []
value = []

for i in obsdict['observations']:
    obs = list(i.values())
    realtime_start.append(obs[0])
    realtime_end.append(obs[1])
    date.append(obs[2])
    value.append(obs[3])
```

```python
value

obs_df = pd.DataFrame(data={cols[0]:realtime_start, cols[1]:realtime_end, cols[2]:date, cols[3]:value}, columns=cols)
obs_df
```
